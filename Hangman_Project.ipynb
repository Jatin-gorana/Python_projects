```python
import random

guess_list = ['Almond', 'Banana', 'Blewits', 'Brazil', 'Carrot', 'Casaba', 'Cashew', 'Celery', 'Cherry', 'Chives',
              'Chocho', 'Citron', 'Cob nut', 'Cooker', 'Dalkon', 'Damson', 'Endive', 'Fennel', 'Garlic', 'Girkin',
              'Kiwano', 'Lentil', 'Lichee', 'Marrow', 'Medlar', 'Nettle', 'Orange', 'Olyster', 'Papaya', 'Pawpaw',
              'Pepper', 'Quince', 'Radish', 'Rocket', 'Russet', 'Squash', 'Tomato', 'Turnip', 'Walnut']

guessing_word = random.choice(guess_list).lower()
word_letters = len(guessing_word)
result = ['_'] * word_letters
attempts = 6
guessed_letters = []

lives = ['''
__________
  |      \|
  o       |
 /|\      |
 / \      |
          |
===========
''', '''
__________
  |      \|
  o       |
 /|\      |
   \      |
          |
===========
''', '''
__________
  |      \|
  o       |
 /|\      |
          |
          |
===========
''', '''
__________
  |      \|
  o       |
 /|\      |
          |
          |
===========
''', '''
__________
  |      \|
  o       |
  |\      |
          |
          |
===========
''', '''
__________
  |      \|
  o       |
  |       |
          |
          |
===========
''', '''
__________
  |      \|
          |
          |
          |
          |
===========
''']

game_name = ''' 

 __
|  |
|  |_____     ____    ______       _____    _________     ____     ______
|   __   |   / _  |  |   _  \     /  _  |  /  _   _  \   / _  |   |   _  \
|  |  |  |  | (_| |  |  | |  |   |  (_| |  | | | | | |  | (_| |   |  | |  |
|__|  |__|  \_____|_ |__| |__|    \___  |  |_| |_| |_|   \____|_  |__| |__|
                                      | |
                                      | |
                                      / / 
                                     /_/

'''

print(game_name)

while attempts > 0 and '_' in result:
    print(' '.join(result))
    print(f'Attempts left: {attempts}')
    print(f"Guessed letters: {' '.join(guessed_letters)}")
    
    guess = input("Guess a letter: ").lower()
    
    if len(guess) != 1 or not guess.isalpha():
        print("Invalid input. Please enter a single letter.")
        continue
    
    if guess in guessed_letters:
        print("You have already guessed that letter. Try again.")
        continue
    
    guessed_letters.append(guess)
    
    if guess in guessing_word:
        for i in range(word_letters):
            if guess == guessing_word[i]:
                result[i] = guess
    else:
        attempts -= 1
        print(f"Wrong guess! Attempts left: {attempts}")
        print(lives[6-attempts])
    
    print()
    
if '_' not in result:
    print(f"The word was {''.join(result)}. You win!")
else:
    print(f"The word was {guessing_word}. You lose!")
```
